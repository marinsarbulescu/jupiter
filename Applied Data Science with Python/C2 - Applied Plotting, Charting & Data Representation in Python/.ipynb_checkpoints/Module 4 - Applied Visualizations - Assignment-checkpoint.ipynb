The direct link to the .xls files on the is not available because the files are generated dynamically, after the user submits a form with different options (https://data.bls.gov/pdq/SurveyOutputServlet is the page where the files are generated).

##### 2. Clean up the data into a dataset with 4 columns
* Year
* National
* LA
* NY

For convenience I've exported the final dataset as "Assig 4 - Unemployment Rate US-LA-NY.csv"

##### 3. Plot the graph showing the correlation of the 3 unemployment trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

# Import .csv
df = pd.read_csv('Resources/Assig 4 - Unemployment Rate US-LA-NY.csv')

# Promote Year column to index
df = df.set_index('Year')

# Define x axis ticks & labels
x_ticks_labels = df.index
x_ticks = df.index

# Wider chart
plt.figure(figsize=(10,5))

# Customize axes
ax = plt.gca()

ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticks_labels)

ax.set_ylabel('% of population')
ax.set_title('City vs National Unemployment Trend (2000 - 2018)')

# Plot the dataset
plt.plot(df['National'], '-', color='gray', linewidth=1.5, alpha=0.3);
plt.plot(df['LA'], '-', color='red', linewidth=1.5, alpha=0.3);
plt.plot(df['NY'], '-', color='blue', linewidth=1.5, alpha=0.3);

# y value intentionally the height of one bar, just so we can show all possible colors
y = 11

# Draw y line
plt.axhline(y = 4, color = 'black', alpha = 0.1, linestyle= '-.', linewidth=0.5)
plt.axhline(y = 8, color = 'black', alpha = 0.1, linestyle= '-.', linewidth=0.5)
plt.axhline(y = 12, color = 'black', alpha = 0.1, linestyle= '-.', linewidth=0.5)

# Add legend
plt.legend(['National', 'Los Angeles', 'New York'])

# Remove frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```
